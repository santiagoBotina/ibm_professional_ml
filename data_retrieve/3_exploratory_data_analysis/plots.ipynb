```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('../data/iris.csv')

data
```

# Basic Scatter plot

```python
# ls='' -> means no lines connecting every dot
# marker='o' -> style for every point(o, x, ...)

plt.plot(data['SepalLengthCm'],
         data['SepalWidthCm'],
        ls='', marker='o')
```

# Scatter Plots with multiple layers

```python
plt.plot(data['SepalLengthCm'],
         data['SepalWidthCm'],
         ls='', marker='o',
         label='sepal')

plt.plot(data['PetalLengthCm'],
         data['PetalWidthCm'],
         ls='', marker='o',
         label='petal')
```

# Customizing Plots

```python
import numpy as np

#fig articulates all the basics around the plot
# ax is the box where the data will be shown

fig, ax = plt.subplots()
ax.barh(np.arange(10),
        data['SepalWidthCm'].iloc[:10])

#Set position of ticks and tick labels
ax.set_yticks(np.arange(0.4,10.4,1.0))
ax.set_yticklabels(np.arange(1,11))
ax.set(xlabel='xlabel', ylabel='ylabel', title='Sepal width')
```

# Grouping data

```python
data.groupby('Species').mean()
```

```python
exluded_id = data.loc[:, data.columns!='Id']

exluded_id.groupby('Species').mean().plot(color=['red','blue','black','green'],
     fontsize=10.0, figsize=(4,4))
```

# Using Seaborn

```python
import seaborn as sns

sns.pairplot(data, hue='Species', height=3)
```

```python
# Hexbin plot -> Shows the relationship and density where most of the values are falling

sns.jointplot(x=data['SepalLengthCm'],
             y=data['SepalWidthCm'],
             kind='hex')
```
